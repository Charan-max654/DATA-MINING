```python
import pandas as pd
```

```python
df = pd.read_csv('/content/employee_performance.csv')
```

```python
print("Employee Performance")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Summary Statistics")
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Monthly_Salary'], bins=10, kde=True, color='blue')
plt.title("Distribution of Sales Amounts")
plt.xlabel("salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Years_of_Experience'], y=df['Performance_Score'], hue=df['Employee_ID'])
plt.title("Years_of_Experience vs. Performance_Score")
plt.xlabel("Years of Experience")
plt.ylabel("Performance Score")
plt.show()
```
